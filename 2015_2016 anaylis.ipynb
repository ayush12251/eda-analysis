```python
student_teacher_ratio= df['tch_6']/df['enr_6']*100
df=df.assign(student_teacher_ratio=student_teacher_ratio)
df[["student_teacher_ratio","enr_6","tch_6"]]
```

```python
literacy_rural_urban=df.melt(id_vars=["statname","literacy_rate"], value_vars=["sch_r_6", "sch_u_6"], var_name="category", value_name="count")
literacy_rural_urban


```

```python
minority_prog=(df["muslim_hsec"]+df["obchsec"]+df["sthsec"]+df["schsec"]).astype(float)/(df["muslim_hsec"]+df["obchsec"]+df["sthsec"]+df["schsec"]+df["scsec"]+df["obcsec"]+df["stsec"]).astype(float)
minority_prog*100
df=df.assign(minority_prog=minority_prog)
```

```python
dropout_after10=df["girls_c10"]-df["girls_c11"]
df=df.assign(dropout_after10=dropout_after10)
df[df["dropout_after10"]<0]=0
dropout_ratio=df["dropout_after10"]/df["girls_c10"]
df=df.assign(dropout_ratio=dropout_ratio)
df["dropout_ratio"].fillna(0, inplace=True)

df["dropout_ratio"]=df["dropout_ratio"]*100
df=df[df["statname"]!="0"]
```

```python
# Ensure correct dtypes for plotting
df['statname'] = df['statname'].astype(str)
cmap = plt.get_cmap("tab10")
df[["water_6","library_6","bndrywall_6"]] = df[["water_6","library_6","bndrywall_6"]].apply(pd.to_numeric, errors='coerce')
fig, ax = plt.subplots(3,2,figsize=(15,15))
sns.lineplot(data=df, x="statname", y="water_6", ax=ax[0,1], label="Water")
sns.lineplot(data=df, x="statname", y="library_6", ax=ax[0,1], label="Library")
sns.lineplot(data=df, x="statname", y="bndrywall_6", ax=ax[0,1], label="Boundary Wall")
ax[0,1].legend()
ax[0,1].tick_params(axis="x", rotation=90)
ax[0,1].set_xlabel("State Name")
plot_df = df.melt(id_vars=["statname"], value_vars=["ict_lab_6", "electric_6"], var_name="facility", value_name="count")
ax[0,1].set_ylabel("Count")
ax[0,1].set_title("facilities vs State")

sns.barplot(data=plot_df, x="statname", y="count", hue="facility", ax=ax[0,0])
ax[0,0].set_xlabel("State Name")
ax[0,0].set_ylabel("Count")
ax[0,0].set_ylim(min(plot_df["count"]) * 0.9, max(plot_df["count"]) * 1.1)
ax[0,0].tick_params(axis="x", rotation=90)
ax[0,0].set_title("ICT Lab vs Electricity")
fig.tight_layout()
color_teacher=["grey" if c>3.3 else "black" for c in df["student_teacher_ratio"]]  # Example colors for student-teacher ratio

sns.barplot(data=df, x="statname", y="student_teacher_ratio", ax=ax[1,0], palette=color_teacher)
ax[1,0].set_xlabel("State Name")
ax[1,0].set_ylabel("Student-Teacher Ratio")
ax[1,0].tick_params(axis="x", rotation=90)
ax[1,0].set_title("Student-Teacher Ratio vs State")
color=[cmap.colors[1] if c<2000 else cmap.colors[2] for c in literacy_rural_urban["count"]]  # Example colors for rural and urban

sns.barplot(data=literacy_rural_urban, x="statname", y="count", ax=ax[1,1], palette=color)
ax[1,1].set_title("Rural vs Urban Literacy Rate")
ax[1,1].set_xlabel("State Name")
ax[1,1].set_ylabel("Count")
ax[1,1].tick_params(axis="x", rotation=90)
ax[1,1].set_title("Rural & Urban Schools vs State")
color_minority=["skyblue" if c>.4 else "lightcoral" for c in df["minority_prog"]]  # Example colors for minority progression


sns.barplot(data=df, x="statname", y="minority_prog", ax=ax[2,0], palette=color_minority)
ax[2,0].set_xlabel("State Name")
ax[2,0].set_ylabel("Minority Progression percentage")
ax[2,0].tick_params(axis="x", rotation=90)
ax[2,0].set_title("Minority Progression vs State")
color_per=["red" if c>50 else "green" for c in df["dropout_ratio"]]  # Example colors for dropout ratio
sns.barplot(data=df, x="statname", y="dropout_ratio", ax=ax[2,1], palette=color_per)
ax[2,1].set_xlabel("State Name")
ax[2,1].set_ylabel("Dropout Ratio percentage")
ax[2,1].tick_params(axis="x", rotation=90)
ax[2,1].set_title("Dropout Ratio of girls after 10th") 
fig.tight_layout()
```
